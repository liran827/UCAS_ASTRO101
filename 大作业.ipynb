# 大作业 撰写一份新一代天文仪器的科学意见书

-所有的同学分为以4-5人为一组的小组, 每组选出一人作为项目科学家

-举行小组会议，进行头脑风暴，列举自己感兴趣的科学问题。对所有的科学进行调查研究，选择3-4项本组认为最为重要的科学目标。

-在列表一中的天文仪器中选择一个作为自己小组的目标仪器

-按照"望远镜口径，视场，空间分辨率，光谱分辨率，指向精度，极限星等"等指标，对本组选择的仪器的能力和适合的科学目标进行分析，优化科学目标，删除不适合的目标，补充可以扩展的目标。

-根据以上讨论结果，撰写一份建造新一代仪器的科学建议书


### 作业提交：

撰写一份科学建议书，需要包括以下内容。

1.对仪器概念设计的描述

2.一份科学要求表(science flowdown chart，见参考文献XXX)，包含每一项科学目标对仪器指标的具体要求。

3. 1-3项核心科学案例列表，需要详述科学案例的重要意义，国际研究现状，以及为什么你选择的仪器，设计的仪器指标可以实现这个科学案例。

4. 加分项：仪器的扩展性讨论，仪器预算估计，仪器建造难点估计，科学发展路线图.....


### 讨论课报告


1. 第二次讨论课，第三次讨论课：介绍世界上现有的天文仪器，包括：设计，核心科学目标，望远镜重要参数，以及望远镜已经达成的最重要科学（选择2项）。每节课2位同学作报告，报告时间各为20分钟。需要在以下几类仪器中选择：通用望远镜（Keck, VLT, TMT), 巡天望远镜(SDSS, LSST), 空间望远镜 (HST, CSST, JWST), 时域巡天望远镜(Kepler, TESS), 射电望远镜 (FAST, ALMA，SKA), 高能望远镜 (Chandra, Fermi, SWIFT)

2.第四，第五次讨论课：介绍各组的科学目标，报告需要包括目标的科学背景和科学意义，并着重阐述什么样的观测设备可以达成这样的目标

3.第六次讨论课：初步报告各自的设计方案，讨论如何制定flowdown chart

4.第七次讨论课：两组同学分别进行大报告，完整论述各自的提议，包括仪器概念，设计，关键指标，主要科学目标

5.第八次讨论课： 由剩下的一位同学，领导讨论地外文明




### 列表一 可供选择的仪器 

1.一个搭载在气球上的望远镜，口径不超过1米，可以包含光谱仪或者多色成像照相机作为后端仪器。

2.一个不超过2米口径的地面望远镜，以时域观测为主。

3.一个搭载在国内12米大光学望远镜上的后端仪器，可以选择多色成像相机， 多目标光纤光谱仪，IFU光谱仪 或者任何你认为合适的仪器。

4.一个搭载在空间站2米望远镜上的小视场精测设备，可以选择多色成像相机，长缝光谱仪，IFU光谱仪。

5.一个不超过50cm 口径的空间望远镜。

6.一个四米口径的巡天专用望远镜。

7.任何超出此列表范围，预算不超过1亿人民币的天文仪器
